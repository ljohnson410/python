```python
import numpy as np
import pandas as pd
```

```python
#Enter output name
outputName = ''
#Enter output location
outputLocation = ''

#imports csv file
df = pd.read_csv(r'C:\Users\Administrator\Desktop\pythonProjects\counties.csv')

df.head()
```

```python
#parses out column names for number fields
NameDict =[]
NumColumns = df.dtypes[(df.dtypes == np.int64) | (df.dtypes == np.float64)]
columnName = list(NumColumns.index)
print(columnName)
```

```python
#appends column name
NameDict.append(columnName)
```

```python
#ranks each number column from highest to lowest and adds new column with ranking

for value in NameDict[0]:
    df['Ranking'+' '+ value] = df[value].rank(axis=0, method='min', numeric_only=True, na_option='bottom',ascending=False, pct=False )
```

```python
df.head()
```

```python
df.to_csv(r'{}{}.csv'.format(outputLocation, outputName), index=False)
```
